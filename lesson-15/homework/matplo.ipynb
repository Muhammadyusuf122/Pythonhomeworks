```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10, 100)
def fx(x):
     return x**2 - 4*x + 4
y = fx(x)
plt.plot(x,y, label ='$ f(x) = x^2 - 4x + 4 $')
plt.title('Quadratic function')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()


a= np.linspace(0, 2*np.pi,100)
sin_a = np.sin(a)
cos_a = np.cos(a)
plt.plot(a, sin_a, c='g', linestyle = '-', marker = 'o', label = ' $ \sin(x) $ and $ \cos(x) $')
plt.plot(a, cos_a, c='b', linestyle = ':', marker = 'o', label = ' $ \sin(x) $ and $ \cos(x) $')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,100)
y1 = x**3
y2 = np.sin(x)
y3 =  np.e**x
y4 = np.log(x+1)
plt.figure(figsize = (10,10))
plt.subplot(2,2,1)
plt.plot(x, y1, c='r', marker = 'o', linestyle = '--', label =' $ f(x) = x^3 $')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Cube function")
plt.legend(loc='upper left')

plt.subplot(2,2,2)
plt.plot(x, y2, c = 'y', marker ='*', linestyle =':', label = '$ f(x) = \sin(x) $')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trigonometric function')
plt.legend(loc = 'upper right')

plt.subplot(2,2,3)
plt.plot(x, y3, c = 'y', marker = 'o', linestyle = '-', label = ' $ f(x) = e^x $')
plt.xlabel('x')
plt.ylabel('y')
plt.title('e value')
plt.legend(loc = 'lower left')

plt.subplot(2,2,4)
plt.plot(x,y4, c = 'skyblue', marker ='*', linestyle = '-.', label = '$ f(x) = \log(x+1) $ (for $ x \geq 0 $')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Logarithmic function')
plt.legend(loc = 'lower right')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)


x = np.random.randint(0,10,100)
y = np.random.randint(0,10,100)
colors = np.random.rand(100)


scatter = plt.scatter(x,y, c=colors, marker= 'D', s=50, label =' Scatterplots', cmap = 'viridis')
plt.title('My Scatterplot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.colorbar(scatter, label = 'Color intensity')
plt.grid(True, linestyle='--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
data = np.random.normal(0,1, 1000)
plt.hist(data, color = 'green', bins = 30, alpha = 0.7, label= 'Histogram', edgecolor='k' )
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('My histogram')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5,5,200)
y = np.linspace(-5,5,200)
x,y = np.meshgrid(x,y)
z = np.cos(x**2 + y**2)
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot: $z = cos(x^2 + y^2)$')

fig.colorbar(surface, ax=ax, aspect=5, label='Z-value')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
categories = ['Category A', 'Category B', 'Category C']
T1 =[20,15,10]
T2 =[35,25,20]
T3=[30,20,25]
T4 = [25,30,15]
plt.bar(categories, T1, color= 'skyblue', label='T1')
plt.bar(categories, T2,bottom = T1,color= 'yellow', label='T2')
plt.bar(categories, T3, bottom =np.add(T1,T2), color= 'red', label='T3')
plt.bar(categories, T4, bottom= np.add(T1, np.add(T2,T3)), color= 'PINK', label='T4')
plt.title('Stacked bar chart')
plt.legend()
plt.show()
```
